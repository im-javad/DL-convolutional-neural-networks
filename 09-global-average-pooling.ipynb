# Global Average Pooling (GAP)

## What is Global Average Pooling (GAP)?

**Global Average Pooling (GAP)** is one of the most commonly used layers in Convolutional Neural Networks (CNNs). It is usually placed at the end of the network, just before the output layer.

The purpose of GAP is to **extract only one value from each Feature Map**. This value is equal to the **average of all values within that Feature Map**.

In simple words:

> GAP summarizes each Feature Map into a single representative value.

## Why Was GAP Introduced?

In older CNN architectures such as **AlexNet** and **VGG**, Fully Connected (FC) layers were used after the convolutional layers.

Problems with Fully Connected layers:

- Huge number of parameters
- High memory consumption
- Higher risk of overfitting
- Longer training time

To overcome these limitations, researchers introduced **Global Average Pooling (GAP)**.

# How GAP Works

Suppose the output of the last convolutional layer is:

```text
7 × 7 × 512
```

Which means:

- Height = 7
- Width = 7
- Number of Feature Maps = 512

With GAP:

For each Feature Map:

```text
7 × 7
```

the average value is computed.

For example:

Feature Map:

```text
1 2 3
4 5 6
7 8 9
```

Average:

```text
(1+2+3+4+5+6+7+8+9)/9

=45/9

=5
```

Therefore, the output of this Feature Map is simply:

```text
5
```
If there are 512 Feature Maps:

```text
7 × 7 × 512

↓

1 × 1 × 512
```

# GAP Formula

If a Feature Map has dimensions

```text
H × W
```

then the GAP output is:

\[
GAP=\frac{1}{H\times W}\sum_{i=1}^{H}\sum_{j=1}^{W}x_{ij}
\]

where:

- H = Height
- W = Width
- \(x_{ij}\) = Value of each element in the Feature Map

# Numerical Example

Suppose we have the following Feature Map:

```text
2 4

6 8
```

Average:

```text
(2+4+6+8)/4

=20/4

=5
```

GAP Output:

```text
5
```

# Another Example

Feature Map:

```text
1 1 2

2 3 4

4 5 8
```

Sum:

```text
1+1+2+2+3+4+4+5+8

=30
```

Number of elements:

```text
9
```

Average:

```text
30/9

=3.33
```

# CNN Architecture with GAP

Before GAP:

```text
Image

↓

Convolution

↓

Pooling

↓

Convolution

↓

Flatten

↓

Fully Connected

↓

Softmax
```

After GAP:

```text
Image

↓

Convolution

↓

Pooling

↓

Convolution

↓

Global Average Pooling

↓

Softmax
```

Notice that **Flatten** and **Fully Connected** layers are no longer required.

# Difference Between GAP and Average Pooling

### Average Pooling

Average Pooling computes the average over a **small local region**.

For example:

```text
Kernel = 2 × 2
```

Input:

```text
1 2 3 4

5 6 7 8

9 1 2 3

4 5 6 7
```

After Average Pooling, the output is still a Feature Map.


### Global Average Pooling

Global Average Pooling computes the average over the **entire Feature Map**.

For example:

```text
7 × 7

↓

1
```

# Difference Between GAP and Max Pooling

### Max Pooling

Max Pooling selects the maximum value within a region.

Example:

```text
1 3

5 2
```

Output:

```text
5
```


### GAP

GAP computes the average of all values.

```text
1+3+5+2

=11

11/4

=2.75
```


# Difference Between GAP and Fully Connected Layer

| Fully Connected | GAP |
|-----------------|-----|
| Large number of parameters | No trainable parameters |
| Higher risk of overfitting | Lower risk of overfitting |
| High memory usage | Low memory usage |
| Slower training | Faster training |

# Advantages of GAP

### 1. Reduces the Number of Parameters

Suppose the convolutional output is:

```text
7 × 7 × 512
```

Using Flatten:

```text
25088
```

inputs are passed to the Fully Connected layer.

Using GAP:

```text
512
```

values are produced.

This dramatically reduces the number of parameters.

---

### 2. Reduces Overfitting

Since trainable parameters are removed, the risk of overfitting decreases.

---

### 3. Faster Training

Fewer parameters mean:

- Faster training
- Lower memory consumption

---

### 4. Simpler Network Architecture

Large Fully Connected layers are no longer necessary.

---

### 5. Well-Suited for Image Classification

Most modern CNN architectures use GAP.


# Disadvantages of GAP

- Some spatial information may be lost.
- GAP alone is not suitable for tasks requiring precise object localization.

# CNN Architectures That Use GAP

- Network in Network (NiN)
- GoogleNet (Inception)
- ResNet
- DenseNet
- MobileNet
- EfficientNet
- Xception

Almost all modern CNN architectures use GAP.

# When Should We Use GAP?

Global Average Pooling is commonly used when:

- The task is **Image Classification**.
- We want to reduce the number of model parameters.
- We want to minimize overfitting.
- We want to replace large Fully Connected layers with a simpler and more efficient solution.
