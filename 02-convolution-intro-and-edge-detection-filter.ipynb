# Convolution in CNNs

Convolution is a mathematical operation used in image processing to extract features. In Convolutional Neural Networks (CNNs), convolution is performed using filters (kernels).

## How it works:
- Consider a 2D image (like a matrix of pixels).
- A small filter (e.g., 3×3) slides over the image.
- At each position, the pixel values and filter values are multiplied and summed to produce a single number.
- The result is called a feature map, highlighting specific patterns in the image.

## Key points:
- Filters can detect edges, corners, textures, or more complex patterns.
- Filter weights are learned during training.

# Edge Detection

Edges are points in an image where there is a sharp change in intensity or color. Detecting edges is important for understanding the structure of an image and extracting features.

### Important note: 
In CNNs, early layers typically detect edges. Later layers combine these features to detect complex shapes and objects.

## In CNNs, convolution is often used for edge detection:
- Specialized edge-detecting filters (like Sobel, Prewitt, or Laplacian) detect changes in pixel intensity in different directions.
- For example, a Sobel filter highlights horizontal or vertical changes.
- The result of convolution with these filters is a new image where edges are bright and other areas are dark.

# Summary Flow

1. Input: raw image
2. Convolutional layer: filters slide over image → feature maps
3. Feature maps: edges and patterns are highlighted
4. Deeper layers: combine feature maps to recognize complex objects
