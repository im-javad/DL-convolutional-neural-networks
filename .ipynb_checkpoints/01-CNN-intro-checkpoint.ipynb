# Into

An MLP network cannot take into account the adjacency of pixels and only sees each pixel as a separate feature, and it's suitable for structured problems, not unstructured examples.

# The Main Problem Before CNNs

Before Convolutional Neural Networks (CNNs), when we tried to feed images or large, complex data into fully connected (dense) networks, there were several problems:
1. High Dimensionality:
    - A simple 256×256×3 (RGB) image has over 197k features (pixels).
    - Dense networks require separate weights for each input → huge number of parameters → heavy computation and need for lots of data.
2. Loss of Spatial Structure:
    - Dense layers ignore the spatial arrangement of pixels.
    - For example, if an object moves in an image, a dense network cannot detect it.
3. Overfitting:
    - With many parameters and limited data, the network memorizes the training set instead of generalizing.

# How CNN Solved These Problems

CNNs introduce convolutional layers and pooling layers to address these issues:
1. Convolution → Reduces Parameters and Preserves Features
    - Each filter (kernel) looks at only a small local region of the image (local receptive field).
    - Weight Sharing: The same filter is applied across the whole image → drastically fewer parameters.
    - Result: The network can detect local patterns (edges, textures, corners) without learning every pixel individually.
3. Pooling → Reduces Dimensions and Adds Translation Invariance
    - MaxPooling or AveragePooling reduces the size of feature maps while keeping important information.
    - Makes the network more robust to object movement in the image.
5. Automatic Feature Learning
    - CNNs don’t require manual feature engineering.
    - Lower layers learn simple features like edges and lines, higher layers learn complex shapes and eventually object-level features.

# Why CNNs Are Used

- Drastically fewer parameters and computations compared to dense networks.
- Ability to detect local and hierarchical patterns in images.
- Robustness to translation and size variations of objects.
- Can learn features automatically without manual engineering.

# Summary

- Problem: Dense networks were inefficient for large images, prone to overfitting, and ignored spatial structure.
- Solution: Convolution, weight sharing, and pooling in CNNs → fewer parameters, better local feature detection, translation invariance.
- Result: CNNs became the standard for image, video, and multidimensional data processing.
