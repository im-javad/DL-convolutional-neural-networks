# Pooling in convolutonal neural networks (CNNs) 

## What's Pooling?

Pooling is a non-learnable operation in Convolutional Neural Networks that is typically applied after convolution layers to reduce the spatial dimensions (width and height) of feature maps while preserving the most important information.

In simple terms: <br/>
Pooling summarizes local regions of a feature map.

## Why is pooling used?

1. dimentionality reduction
    - reduces width and height
    - lowers memory usage
2. computational efficiency
    - fewer operations in deeper layers
3. overfitting reduction
    - less parameters and redundancy
4. translation invariance
    - small shifts in input don't drastically effect output 

## How pooling works?

Pooling operates independently on each feature map.

#### Steps:
1. Select a local window (e.g., 2×2)
2. Slide the window using a stride
3. Apply a pooling function (e.g., max or average)

The number of channels remains unchanged

## Pooling parameters

Kernel size : Size of the pooling window

Stride: Step size of the window movement

Padding: Usually zero

Pooling type: Max, Average, Global

## Output Size Formula

#### Given:
    - input size: W * H
    - kernel size: F
    - stride: S
    - Padding: P 

Output spatial dimension: output size = [(w - F + 2P) / S] + 1

## Types of pooling: 

### Max pooling:

Selects the maximum value within each window.

characteristics: 
- Preserves strongest activations
- Most commonly used pooling method
- Robust to noise and small shifts

### Average pooling:

Computes the average value of the window.

Characteristics:
- Produces smoother feature maps
- Less sensitive to outliers
- Less common in modern CNNs

## Advantages and Disadvantages

Advantages: 
- Reduces spatial dimensions
- Improves generalization
- Provides translation invariance

Disadvantages:
- Loss of precise spatial information
- Not suitable for tasks requiring localization (e.g., segmentation)
