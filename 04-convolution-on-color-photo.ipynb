# Structure of a Color Image (RGB)

### A color image is composed of three channels:
- Red (R)
- Green (G)
- Blue (B)

If the spatial size of the image is: <br/>
Height × Width

Then the full shape of a color image is: <br/>
Height × Width × 3

Example: <br/>
224 * 150 * 3

# Convolution Filters (Kernels) for Color Images

### Key Rule:
In CNNs, the depth of a filter must match the depth of the input image.

So, for an RGB image: <br/> 
Each convolution filter has 3 channels

Filter shape: <br/>
k × k × 3

Example: <br/>
3 × 3 × 3

# How Convolution Works on Color Images

### Assume:
- Input image: H × W × 3
- Filter: k × k × 3

### Steps:
- The filter slides across the image spatially
- At each position:
    - Element-wise multiplication is performed separately on each channel
    - The results from all channels are summed
- A single scalar value is produced for that position
- Mathematically:
    - Output = Σ(R ⊙ KR) + Σ(G ⊙ KG) + Σ(B ⊙ KB)
        - Where:
            - ⊙ denotes element-wise multiplication
            - KR , KG , KB are the filter weights for each channel

# Output of Convolution on Color Images

***One filter → one feature map***

If: 
- 1 filter is used
    - Output shape:
        - H' × W' × 1

If:
- N filters are used
    - Output shape:
        - H' × W' × N

# Numerical Example

- Input:
    - 32 * 32 * 3
- Filters (in first hidden layer)
    - 10 filters of size 5 * 5 * 3
- Output:
    - 28 * 28 * 10
    - 35 - 5 + 1(baias) = 28 (no padding, stride = 1)
