# A sample of calculation (continue previous section) 

## How calculate parameters in con layers?

### ( (f(h) * f(w) * c ) + b ) * nf

f = filter size <br/>
c = number of channels (the filter size takes the number of channels from the previous layer) <br/>
b = bias <br/> 
nf = number of filters 

## How calculate pooling?

activation shape = half fs <br/>
parameters number = 0 ( because it has no weight or bias) 

## How Calculate Fully Connections? 

activation shape = number of neurons <br/>
parameters = number of neurons * previouse activations size + number of neurons bias (equal) 

## How Calculate Softmax? 

activation shape = number of classes <br/> 
parameters = number of classes * previuse layer neurons + number of classes bias (equal)
