# Summary (what's Padding?):
Padding is the process of adding extra pixels (usually zeros) around the border of the input before applying convolution. <br/>
It is used to control output size, preserve edge information, and improve feature extraction.

# Problems before using Padding:
1. Output size shrinks after each convolution
    - 5×5 input → 3×3 output (with 3×3 kernel)
2. Loss of edge information
    - Edge pixels are convolved fewer times than center pixels.
3. Asymmetry
    - Central pixels get a full receptive field; border pixels do not.
4. Deep networks collapse in size
    - After many layers the feature maps become very small.

# Why do we use Padding?
- To preserve the spatial dimensions (output same size as input).
- To avoid losing edge information.
- To maintain symmetry in how all pixels are processed.
- To allow building deep architectures without constant shrinking.
- To improve feature extraction by providing more valid positions for kernels.

# What Padding Depends On
1. Kernel size
    - Larger kernels → more padding needed.
2. Stride
    - When stride > 1, padding may need to increase.
3. Desired output size
    - Same size → use same padding
    - Smaller size → use valid padding
4. Network architecture requirements
5. Framework conventions (TensorFlow vs PyTorch)

# Stride
Stride in CNNs is the number of pixels the convolution filter moves at each step across the input.
- Stride = 1: Filter moves one pixel → larger output, more detail.
- Stride > 1: Filter jumps more pixels → smaller output, more downsampling.
- Bigger stride = fewer computations and less overlap.
